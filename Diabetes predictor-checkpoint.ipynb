# Data collection and analysis

### Loading and reading the data set

```python
import pandas as pd
diabetes_dataset = pd.read_csv('diabetes.csv')
diabetes_dataset.head()
```

```python
#getting the number of rows and columns
diabetes_dataset.shape
```

```python
#getting statistics of the data
diabetes_dataset.describe()
```

```python
#Seeing how many people had diabetes and how many didn't
# 0 -> Non diabetic, 1 ->diabetic
diabetes_dataset['Outcome'].value_counts()
```

```python
diabetes_dataset.groupby('Outcome').mean()
```

### Separating the data and labels (input and output)

#### Input (X) 

```python
X = diabetes_dataset.drop(columns = 'Outcome', axis=1)
X
```

#### Output Y

```python
Y = diabetes_dataset['Outcome']
Y
```

# Data Standerdization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
standerdized_data = scaler.transform(X) #storing the standerdized data to a variable

standerdized_data
```

```python
X = standerdized_data #returning the standerdized data back to variable X

X
Y
```

# Seperating tran and test data

### Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2) 

print(X.shape, X_train.shape, X_test.shape)
```

### Training the model

```python
from sklearn import svm
classifier = svm.SVC(kernel = 'linear')
```

```python
#Training the support vector machine classifier
classifier.fit(X_train, Y_train)
```

# Testing the accuracy

### Evaluating the accuracy score on training data

```python
from sklearn.metrics import accuracy_score
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

training_data_accuracy
```

### Evaluating the accuracy score on test data

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

test_data_accuracy
```

# Making a predictive system

```python
import numpy as np
input_data = (1,85,66,29,0,26.6,0.351,31)

#change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

#standerdizing the input data
std_data = scaler.transform(input_data_reshaped)
std_data

prediction = classifier.predict(std_data)

if prediction[0] == 0:
    print ('The patient does not have diabetes')
    
else:
        print('The person is diabetic')
```
